```python
# coding=utf-8
from __future__ import print_function, absolute_import, unicode_literals

import MySQLdb
import codecs
from datetime import *
import time
from DB import *
import json

# 20100506 -> 2010-05-06
def mysqldate(str):
    return str[0:4]+"-" + str[4:6] + "-" + str[6:]

db = DB()
n = 0
with open("/Users/abing/Downloads/index_components_399932.XSHE.csv", "r") as fin:
    date_code = fin.readline()
    constituents = fin.readline()
    while date_code:
        (date, code) = date_code.split()
        code = code.split(".")[0]
        date = mysqldate(date)
        
        print(date, code)
        sql = "select trade_date from index_constituents where code = '%s' and trade_date <= '%s' order by trade_date desc limit 1" % \
        (code, date)
        
        result = db.executeSql(sql)
 
        constituents = json.loads(constituents)
        
        no_weight = 0
        if type(constituents) == "dict":
            adjusted_con = {}
            for key in constituents.keys():
                if key[0] == '6' or key[0] == '9':
                    k = "SHSE."+key[0:6]
                else:
                    k = "SZSE."+key[0:6]

                adjusted_con[k] = constituents[key]*100
        else:
            adjusted_con = []
            no_weight = 1
            for key in constituents:
                if key == '6' or key == '9':
                    adjusted_con.append("SHSE."+key[0:6])
                else:
                    adjusted_con.append("SZSE."+key[0:6])

        
        needinsert = False
        if len(result) > 0:
            print(result[0][0])
            latest = result[0][0]
            timedelta = datetime.strptime(date, '%Y-%m-%d').date() - latest
            print(timedelta,timedelta.days)
            if timedelta.days > 5:
                needinsert = True
            
        if len(result) == 0 or needinsert:
            constituents = json.dumps(adjusted_con)
            sql = "insert into index_constituents (code, trade_date, constituents, no_weight) values ('%s', '%s', '%s', %d)" % \
            (code, date, constituents, no_weight)
            db.executeSql(sql)
            print(sql)
        
        date_code = fin.readline()
        constituents = fin.readline()
        n+=1
#         if n > 5:
#             break

# print(mysqldate("20100607"))

# db.getStockList()
```
